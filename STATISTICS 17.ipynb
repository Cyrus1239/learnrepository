# Categorical variable test of independence----chi square test

```python
# checking the chi square test
```

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
ds=pd.read_csv('role_website.csv')
```

```python
df=pd.DataFrame(ds)
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df.dtypes
```

```python
# create contigency table
df_crosstab=pd.crosstab(df['role'],df['website'],margins=True)
print(df_crosstab)
```

```python
df_crosstab.columns=['facebook','linkedin','row_totals']
df_crosstab.index=['student','teacher','col_totals']
df_crosstab
```

```python
# Get the observed table
observed=df_crosstab.iloc[0:2,0:2]
observed
```

```python
#st.chi2_contingency?
```

```python
"""Methodology 2 : Calculate chi square using scipy.stat library"""

chi2,p,dof,expected=stats.chi2_contingency(observed=observed)

# you can check the return data by calling it
# The returned data includes: chi_squared_stat,p_value,df,expected_crosstab
# print('stresult=',stresult)

print('Chi squre test=',chi2)
print('P-value=',p)
print('dof=',dof)
print('expected=',expected)
"""from this we interpret the p-value is .016 that is less than .05--> significant value.
hence null hypothesis is rejected and shows the two variables are dependent of each other."""
```

# ANOVA TEST

Analysis of Variances

Anova test used to compare the means more than 2 groups (t-test can be used to compare groups)
Group mean differences inferred by analyzing variances
ANOVA uses variance-based F test to check the group mean equality.
one-way (one factor) and two-way (two factors) ANOVA (factor is an independent variable)

ANOVA Hypothesis


Null hypothesis: Groups means are equal (no variation in means of groups)
H0: 1=2==p

Alternative hypothesis: At least, one group mean is different from other groups
H1: All  are not equal

```python
How ANOVA works?
Check sample size : equal number of observation in each group
calculate Mean Square for each group (MS) (SS of group/level-1); level-1 is a degree of freedom (df) for a group
calculate mean Square error (MSE) (SS error/df of residuals)
Calculate F value (MS of group/MSE)
```

```python
# less find the analysis of variances (ANOVA) among marks conducted for maths test in three section.
# The below is given marks list of three section

from scipy.stats import f_oneway

Marks_SectionA=[88,90,20,30,45]
Marks_SectionB=[50,60,70,44,33]
Marks_SectionC=[70,80,90,30,20]

f_oneway(Marks_SectionA,Marks_SectionB,Marks_SectionC)
```

# Discover outliers with mathematical function

Z-score-

Wikipedia Definition The Z-Score is the signed number of standard deviation by which the value of an observation or data point is above the mean value of what is being observed or measured

The Instution behind Z-score is to describe any data point by finding their relationship with the standard Deviation and mean of the groupof data points

Z-score is finding the distribution of data where mean is 0  and standard deviation is 1 i.e. normal distribution
we must be wondering that, how does this help in identifying the outliers?
Well, while calculating the Z-score we re-scale and center the data and look for data points which are too far from zero.
These data points which are way too far from zero will be treated as the outliers.
in most of the case a threshold of 3 or -3 is used i.e.if the z-score value is greater than or less than 3 or -3 respectively, that data point will be identified  as outliers.


```python
from scipy.stats import zscore
import numpy as np
```

```python
z=np.abs(zscore(hf))
```

```python
z
```

Looking the code and the output above. It is diffucult to say which data point is an outlier. Let's try and define a threshold to indentify an outlier.

```python
threshold=3
print(np.where(z>3))
```

```python
# array 1 is holding the row number and array 2 is holding the column
# all the values are greater than 3 means all are outlier
# Z[Row][Column]
z.iloc[53][0]
```

```python
z.iloc[1457][2]
```

In the previous section, we saw how one can detect the outlier using Z score but now we want to remove or filter the outliers and get the clean data. This can be done with just one line code as we have already calculated the Z score

```python
hf_new=hf[(z<3).all(axis=1)]
hf_new
```

```python
hf.shape
```

```python
hf_new.shape
```

So, above code removed around 100+ rows from the datasheet i.e. outliers have been reomved.
